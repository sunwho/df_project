```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
df_x = df[['흡연자밀도','유동인구수','거주인구수','민원수','담배소비금액','상권발달정도','현재실외흡연실개수']]
# VIF 계산 함수
def calculate_vif(df):
    vif_data = pd.DataFrame()
    vif_data["feature"] = df.columns
    vif_data["VIF"] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    return vif_data

# VIF 계산 함수 실행
vif_results = calculate_vif(df_x)
print(vif_results)
```

```python
from pyproj import Transformer
from shapely.geometry import Polygon
from shapely.wkt import loads, dumps

# TM128 좌표계에서 WGS84(위도/경도)로 변환기 설정
transformer = Transformer.from_crs("epsg:5179", "epsg:4326", always_xy=True)

# 좌표 변환 함수 정의 (TM128 -> WGS84)
def convert_polygon_to_wgs84(polygon_tm):
    try:
        # POLYGON 객체로 변환 (WKT 형식)
        polygon = loads(polygon_tm)
        
        # POLYGON 내 각 좌표를 위도/경도로 변환
        transformed_coords = [transformer.transform(x, y) for x, y in polygon.exterior.coords]
        
        # 변환된 좌표로 새로운 POLYGON 반환
        return Polygon(transformed_coords)
    
    except Exception as e:
        print(f"Error in converting polygon: {e}")
        return None

# 예시 POLYGON 데이터 (주신 데이터 일부 예시)
polygon_tm = """POLYGON ((197702.0691306734 553187.3114751425, 197703.4313579294 553171.653652382, 197704.57146287934 553158.2958395035, ...))"""  # 실제 데이터에서 일부 생략

# 좌표 변환 실행
polygon_wgs84 = convert_polygon_to_wgs84(polygon_tm)

# 변환된 POLYGON 확인
print(dumps(polygon_wgs84, rounding_precision=6))
```

```python
from pyproj import Proj, transform


# TM128 좌표계를 WGS84 (위도/경도)로 변환
# TM128: EPSG:5179, WGS84: EPSG:4326
proj_tm128 = Proj(init='epsg:5179')  # TM128 좌표계
proj_wgs84 = Proj(init='epsg:4326')  # WGS84 (위도/경도)

# 좌표 변환 함수 정의
def convert_to_wgs84(x, y):
    lon, lat = transform(proj_tm128, proj_wgs84, x, y)
    return lon, lat

# 좌표 변환 적용 (예: 'x_col', 'y_col'이 x, y 좌표를 담고 있다고 가정)
df_seoul_map[['longitude', 'latitude']] = df_seoul_map.apply(
    lambda row: convert_to_wgs84(row['x_col'], row['y_col']), axis=1, result_type='expand'
)

# 변환된 데이터 확인
print(df_seoul_map[['longitude', 'latitude']])
```
